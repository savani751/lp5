```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load detaset
fahion_mnist= keras.datasets.fashion_mnist
```

```python
(train_imgs, train_lbls), (test_imgs, test_lbls) = fahion_mnist.load_data()
```

```python
train_imgs = train_imgs/255
test_imgs = test_imgs / 255
```

```python
#define Model
model = keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),
                         keras.layers.Dense(128, activation='relu'),
                         keras.layers.Dense(10, activation='softmax')
                         ]
)
```

```python
#compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
```

```python
model.fit(train_imgs, train_lbls, epochs=10)
```

```python
test_loss, test_acc = model.evaluate(test_imgs, test_lbls)
print("testLoss: ", test_loss, "Test axx: ", test_acc)
```

```python
#prediction

test_pred = model.predict(test_imgs)
predicted_labels = np.argmax(test_pred, axis= 1)
```

```python
L = 5
W = 5

fig,axes = plt.subplots(L,W,figsize = (12,12))
axes = axes.ravel()
for i in np.arange(0,L*W):
    axes[i].imshow(test_imgs[i])
    axes[i].set_title('Prediction Class:{1} \n true class: {1}'.format(predicted_labels[i],test_lbls[i]))
    axes[i].axis('off')
plt.subplots_adjust(wspace = 0.5) 
```
